```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from lininterp import interp1d as ip
```

```python
a, b = 0, 4

f = lambda x: x**2
```

```python
n1, n2 = 5, 100
coarse_grid = np.linspace(a, b, n1)
fine_grid = np.linspace(a-2, b+2, n2)
```

```python
func_vals = f(coase_grid)
```

```python
fig, ax = plt.subplots(figsize=(9, 6))
ax.plot(fine_grid, f(fine_grid))
ax.plot(fine_grid, ip(coarse_grid, func_vals, fine_grid))
```
